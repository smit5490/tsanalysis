## tsanalysis Examples

This notebooks demonstrates the usage of the tsanalysis module.

```python
# Import module
import tsanalysis as ts

# Import other necessary modules
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

### Create dataframe

```python
sin_input = np.linspace(0,2*math.pi*36, num = 365)
val_1 = np.array([math.sin(val) for val in sin_input])
dt = pd.date_range("01-01-2020", periods = 365)

df = pd.DataFrame({"dt": dt,"val_1": val_1})
```

```python
df.head()
```

### Instantiate object and print attributes

```python
tsdf = ts.Tsdf(df, dt_column = "dt")
```

```python
tsdf.df.head() # Dataframe is available as an attribute
```

```python
tsdf.dt_column # name of datetime column
```

```python
tsdf.max_date # max value of datetime column
```

```python
tsdf.min_date # min value of datetime column
```

### tsanalysis methods

```python
tsdf.find_gaps(freq = "1 day") # Check for gaps in time-series
```

```python
# Add time-series decomposition columns to df
tsdf.ts_decomp(components = ["year", "month", "day", "hour"]) 
```

```python
tsdf.df.head()
```

### tsanalysis plots

```python
tsdf.ts_plot(y = "val_1")
```

```python
tsdf.ts_plot(y = "val_1", 
             start_time = "2020-03-01", 
             end_time = "2020-03-15")
```

```python
tsdf.ts_heatmap(fill = "val_1", x_val = "day", y_val = "month")
```
